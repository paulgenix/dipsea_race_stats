```python
import pandas as pd
```

```python
df1 = pd.read_csv('/Users/paul/Desktop/Dipsea/PROCESSED/1905-1999/merged/merged_processed_1905to1999.csv')
df2 = pd.read_csv('/Users/paul/Desktop/Dipsea/PROCESSED/2000-2016/merged/merged_processed_2001to2016.csv')
df3 = pd.read_csv('/Users/paul/Desktop/Dipsea/PROCESSED/2017-2024/merged/merged_processed_2017to2024.csv')
```

```python
df1.rename(columns={'Year': 'year'}, inplace=True)
df1.drop(columns=['race_place'], inplace=True)
df1.head()
```

```python
#moving columns around
df2 = df2[['place_overall', 'name', 'time', 'actual_time', 'age', 'sex', 'year']]
df2.rename(columns={'time': 'clock_time'}, inplace=True)
df2.head()
```

```python
df3['name'] = df3['first_name'] + ' ' + df3['last_name']
df3 = df3.drop(columns=['first_name', 'last_name', 'actual_place'])
df3 = df3[['place_overall', 'name', 'clock_time', 'actual_time', 'age', 'sex', 'year']]
df3.head()
```

```python
# Now, all DataFrames have the same structure
# Combine them into one DataFrame
combined_df = pd.concat([df1, df2, df3], ignore_index=True)
combined_df.to_csv('/Users/paul/Desktop/Dipsea/PROCESSED/allyears_processed.csv', index=False)
```
