```python
import pandas as pd
import glob

folder_path = '/Users/paul/Desktop/Dipsea/RAW/1905-1999'
csv_files = glob.glob(f"{folder_path}/*.csv")

dfs = []
for file in csv_files:
    year = int(file.split('_')[-1].split('.')[0])
    
    df = pd.read_csv(file)
    
    df['Year'] = year
    
    dfs.append(df)

if dfs:
    combined_df = pd.concat(dfs, ignore_index=True)
    

    combined_df.head()
else:
    print("No CSV files")
```

```python
combined_df.sort_values(by='Year', ascending=True, inplace=False)
```

```python
combined_df.to_csv('/Users/paul/Desktop/Dipsea/PROCESSED/1905-1999/combined_data_1905-1999.csv', index=False)
```
